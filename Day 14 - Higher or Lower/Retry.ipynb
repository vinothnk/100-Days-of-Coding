```python
from art import logo
from art import vs
from game_data import data

import random

#1st - pull random data from dict
def generate_data():
    return random.choice(data)





def check_same():
    if A == B:
        B = generate_data()
    return


def main():
    print(logo)
    score = 0
    game_start = True
    
    A = generate_data()
    print(A)
    print(vs)
    B = generate_data()
    print(B)
    if A == B:
        B = generate_data()
        
    while game_start: 
        if A == B:
            B = generate_data()
        else:  
            question = input("Which is higher, A or B?: ").lower()
        
            if question == "a":
                if A["follower_count"] > B["follower_count"]:
                    score += 1
                    print(f"Your score is {score}.")
                    A = B
                    B = generate_data()
                    print(A)
                    print(vs)
                    print(B)
                    game_start = True
                else:
                    print(f"Your score is {score}.")
                    print("Game Over.")
                    game_start = False
            elif question == "b":
                if A["follower_count"] < B["follower_count"]:
                    score += 1
                    print(f"Your score is {score}.")
                    A = B
                    B = generate_data()
                    print(A)
                    print(vs)
                    print(B)
                    game_start = True
                else:
                    print(f"Your score is {score}.")
                    print("Game Over.")
                    game_start = True
            
main()
```

```python
import random
from art import logo
from art import vs
print (logo)

from game_data import data

#generate 1st details
def generate_A():
    global name_A
    global description_A
    global country_A
    global follower_count_A
    num_A = random.randint(0,49)
    name_A = data[num_A]['name']
    description_A = data[num_A]['description']
    country_A = data[num_A]['country']
    follower_count_A = data[num_A]['follower_count']
    print(name_A)
    print(f"follower count is {follower_count_A}")
    return

#generate 2nd details
def generate_B():
    global name_B
    global description_B
    global country_B
    global follower_count_B
    num_B = random.randint(0,49)
    name_B = data[num_B]['name']
    description_B = data[num_B]['description']
    country_B = data[num_B]['country']
    follower_count_B = data[num_B]['follower_count']
    print(name_B)
    print(f"follower count is {follower_count_B}")
    return

def change_B_to_A():
    name_A = name_B
    description_A = description_B
    country_A = country_B
    follower_count_A = follower_count_B

generate_A()
generate_B()

print(f"Compare A: {name_A}, {description_A}, {country_A}")
print(vs)
print(f"Compare B: {name_B}, {description_B}, {country_B}")

question = input("Which category has more followers? A/B?: ")
question = question.upper()

score = 0

if follower_count_A > follower_count_B and question == "A":
    global score
    score += 1
    print(f"Your score is {score}.")
    change_B_to_A()
    print(f"Compare A: {name_A}, {description_A}, {country_A}")
    print(vs)
    print(f"Compare B: {name_B}, {description_B}, {country_B}")
elif follower_count_A < follower_count_B and question == "A":
    print(f"Your score is {score}.")
    print("Game Over.")
elif follower_count_A > follower_count_B and question == "B":
    print(f"Your score is {score}.")
    print("Game Over.")
elif follower_count_A < follower_count_B and question == "B":
    score += 1
    print(f"Your score is {score}.")
    change_B_to_A()
    print(f"Compare A: {name_A}, {description_A}, {country_A}")
    print(vs)
    print(f"Compare B: {name_B}, {description_B}, {country_B}")
```
